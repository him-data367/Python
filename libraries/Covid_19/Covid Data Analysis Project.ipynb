#### Coronavirus disease (COVID-19) is an infectious disease caused by the SARS-CoV-2 virus.Most people infected with the virus will experience mild to moderate respiratory illness and recover without requiring special treatment. However, some will become seriously ill and require medical attention. Older people and those with underlying medical conditions like cardiovascular disease, diabetes, chronic respiratory disease, or cancer are more likely to develop serious illness. Anyone can get sick with COVID-19 and become seriously ill or die at any age.

```python
covid_data.drop(columns=['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'],inplace=True)
```

```python
covid_data['Date']=pd.to_datetime(covid_data['Date'])
```

```python
covid_data.head()
```

##### Active case

```python
covid_data['Active cases']=covid_data['Confirmed']-covid_data['Cured']+ covid_data['Deaths']
```

```python
covid_data.head()
```

##### Pivot table

```python
statewise = pd.pivot_table(covid_data,values=['Confirmed','Deaths','Cured'],index = 'State/UnionTerritory',aggfunc=max)
```

##### Recovery and Mortality rate

```python
statewise['Recovery rate'] = statewise['Cured']*100/statewise['Confirmed']
statewise['Mortality rate'] = statewise['Deaths']*100/statewise['Confirmed']
```

```python
statewise = statewise.sort_values(by='Confirmed',ascending=False)
```

##### Create  a pivot chart

```python
statewise.style.background_gradient(cmap='GnBu')
```

##### Top 10 Active cases states

```python
top_10_active_cases=covid_data.groupby("State/UnionTerritory").max()[['Active cases','Date']].sort_values(by='Active cases',ascending=False).reset_index()
```

```python
plt.figure(figsize=(16,9))
plt.title('Top 10 Most Active cases in India',size=25)
ax =sns.barplot(data = top_10_active_cases.iloc[:10], y ='Active cases',x='State/UnionTerritory',linewidth=2,edgecolor='black')
plt.xlabel('States',size=15)
plt.ylabel('Total Active cases',size=15)
```

##### Top States with Highest Death

```python
top_10_death = covid_data.groupby('State/UnionTerritory').max()[['Deaths','Date']].sort_values(by ='Deaths',ascending = False).reset_index()
```

```python
plt.figure(figsize=(18,5))
plt.title("Top 10 State With Most Death",size =25)
ax= sns.barplot(data=top_10_death.iloc[:12],y='Deaths',x='State/UnionTerritory',linewidth=2,edgecolor='black')
plt.xlabel("States",size=15)
plt.ylabel('Total Death Cases',size=15)
plt.show()
```

##### Growth Trend

```python
plt.figure(figsize=(12,6))
ax = sns.lineplot(data = covid_data[covid_data['State/UnionTerritory'].isin(['Maharashtra','Karnataka','Tamil Nadu','Delhi',
                                                                            'Uttar Pardesh','West Bengal','Kerela','Punjab',
                                                                             'Andhra Pardesh','Chattisgarh'])],x='Date',y='Active cases',hue='State/UnionTerritory')
ax.set_title("Top 5 affected State in India",size=16)
                  
                                    
```

#####  Remove rows where state = India`

```python
vaccine = vaccine_data[vaccine_data.State!='India']
```

```python
vaccine.head(2)
```

```python
vaccine.rename(columns={"Total Individuals Vaccinated":"Total"},inplace=True)
```

```python
vaccine.head(2)
```

#####  Find state vaccinated state

```python
max_vaccine = vaccine.groupby('State').sum()
max_vaccine = max_vaccine.sort_values(by='Total',ascending=False)[:5]
max_vaccine.head(2)
```

```python
plt.figure(figsize=(10,5))
plt.title("Top 5 Vaccinated States in India",size=20)
x = sns.barplot( data= max_vaccine.iloc[:10],y =max_vaccine.Total, x=max_vaccine.index ,linewidth=2,edgecolor='black')
plt.xlabel('States',size=12)
plt.ylabel('Vaccination',size=12)
plt.show()
```

```python
min_vaccine = vaccine.groupby('State').sum()
min_vaccine = min_vaccine.sort_values(by='Total',ascending=True)[:5]
min_vaccine.head(2)
```

```python
plt.figure(figsize=(14,5))
plt.title("Top 5 Vaccinated States in India",size=20)
x = sns.barplot( data= min_vaccine.iloc[:10],y =min_vaccine.Total, x=min_vaccine.index ,linewidth=2,edgecolor='black')
plt.xlabel('States',size=12)
plt.ylabel('Vaccination',size=12)
plt.show()
```
